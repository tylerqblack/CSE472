```python
with open("./datasets/Un_crawled_data/unlabeled.tsv", "r") as f :
    TSV = f.readlines()
print(len(TSV))    

classes = {"human":True, "bot":False}
dataset = []
for label in TSV:
    #All user_ids in TSV are bot.
    user_id = label.split()[0]
    label = label.split()[1]
    dataset.append([user_id, classes[label]])

print(dataset[0])
```
